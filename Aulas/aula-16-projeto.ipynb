```python
def get_datasets(y, labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'label': label[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return {
        'title': title,
        'display': display
    }
```

```python
def create_chart(x, y, labels, kind='bar',title=''):
    
    datasets = get_datasets(y, labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
        'options': options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    str_aux = f'{url_base}?c={str(chart)}'
    resp = r.get(str_aux)
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_PIL = Image.open(path)
    display(img_PIL)
```

```python
y_data_1 = []
for obs in final_data[1::100]: #ignora cabealho e pega dades a cada 100 linhas
    y_data_1.append(obs[confirmados])
    
y_data_2 = []
for obs in final_data[1::100]: #ignora cabealho e pega dades a cada 100 linhas
    y_data_2.append(obs[recuperados])
    
labels = ['Confirmados', 'Recuperados']

x = []

for obs in final_data[1::100]: #ignora cabealho e pega dades a cada 100 linhas
    x.append(obs[data].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Grfico confirmados vs recuperados')
chart_content = get_api_chart(chart)
save_image('meu-primeiro-grafico.png', chart_content)
display_image('meu-primeiro-grafico.png')
```

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
    text = quote(link) # parsing do link para url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```
